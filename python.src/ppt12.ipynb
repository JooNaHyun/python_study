```python
f=0 #파일 객체를 나타내기 위한 변수 f를 초기화
try: #예외가 발생할 가능성이 있는 코드를 실행
    f=open("LICENSE.txt","rt") 
    #현재 디렉토리에 있는 LICENSE.txt 파일을 열고, 파일 객체 f를 생성
    while True: #무한 루프
        r=f.readline() 
        # 파일 객체 f에서 한 줄씩 읽어와 변수 r에 저장
        if not r:
            break
            # 읽어온 데이터가 없다면, 즉 파일의 끝에 도달하면 무한 루프를 종료
    print(r)
    #읽어온 한 줄을 출력
except FileNotFoundError: 
    print("파일없음")
    #LICENSE.txt 파일을 찾을 수 없는 경우 "파일없음"을 출력
finally:
#예외 발생 여부와 상관없이 실행
    if(f): f.close() 
    #파일 객체 f가 열려있다면, 파일을 닫는다.
    print("프로그램 끝")
    #프로그램이 끝났음을 알리는 문자열을 출력
```

```python
import random as r
#파이썬 내장 모듈인 random을 r이라는 이름으로 import
f = open("score.txt", "wt")
#현재 디렉토리에 score.txt 파일을 쓰기 모드로 열어서 파일 객체 f에 저장
for i in range(100):
# 0부터 99까지의 수를 반복
    one_person = "홍길동" + str(i) + "," + str(r.randrange(0, 100)) + "," + str(r.randrange(0, 100)) + "," + str(r.randrange(0, 100)) + "\n"
    #이름이 홍길동이고, 각 과목의 점수가 0에서 99까지 랜덤으로 생성되는 문자열 one_person을 만든다. 
    #마지막에 개행문자(\n)를 붙여서 파일에 쓰여질 때 한 줄씩 쓰이도록 한다.
    f.write(one_person)
    #one_person에 저장된 문자열을 파일에 쓴다.
    print(one_person)
    #one_person에 저장된 문자열을 화면에 출력
f.close()
#파일 객체 f를 닫는다.
```

```python
f = 0 # 변수 f를 초기화
f = open("score.txt", "rt") # score.txt 파일을 읽기 모드로 열기
while True: # 무한루프
    one_person = f.readline() # 파일에서 한 줄씩 읽기
    if not one_person: # 읽을 내용이 없으면
        break # 반복문 탈출
    one_person = one_person.replace("\n", "") # 읽은 문자열에서 \n 제거
    item = one_person.split(",") # 읽은 문자열을 콤마(,)로 분리하여 리스트로 만듦
    print("="*10) # 구분을 위한 줄 출력
    print("이름: ", item[0]) # 학생 이름 출력
    print("국어: ", item[1]) # 국어 성적 출력
    print("영어: ", item[2]) # 영어 성적 출력
    print("수학: ", item[3]) # 수학 성적 출력
    print("="*10) # 구분을 위한 줄 출력
f.close() # 파일 닫기
```

```python
import sqlite3 as s
#모듈을 s라는 이름으로 import한
con = s.connect("my01.db")
#파일과 연결하여 데이터베이스에 접속
cursor = con.cursor()
#접속한 데이터베이스와 상호작용하기 위한 cursor 객체를 생성
cursor.execute("drop table if exists examtable;")  # table이 존재하면 삭제 
cursor.execute(""" 
    create table examtable (
        name varchar(20),
        studentid int not null primary key,
        kor int,
        eng int,
        mat int
    )
""")
#"""을 사용하여 여러 줄의 SQL 쿼리를 작성
#examtable 테이블을 생성
#execute() 함수를 사용하여 SQL 쿼리를 실행시킵니다..
#create table을 사용하여 새로운 테이블을 생성합니다.
#examtable은 테이블 이름입니다.
#name, studentid, kor, eng, mat은 테이블의 컬럼(Column) 이름입니다.
#varchar(20)은 문자열 타입으로 최대 20자까지 저장 가능한 데이터 타입입니다.
#int는 정수형 데이터 타입입니다.
#not null은 해당 컬럼이 null 값을 허용하지 않는다는 것을 의미합니다.
#primary key는 해당 컬럼을 테이블의 기본 키(Primary Key)로 설정
# kor,eng,mat 변수를정수혀으로 설정
con.commit()   # 버퍼에 처리된 내용을 기록(commit)
cursor.close() # cursor 객체를 닫는다
con.close() #데이터베이스 접속을 닫는다.
```

```python
import sqlite3 as s 
con=s.connect("my01.db")  # "my01.db"라는 이름의 데이터베이스와 연결
cursor = con.cursor()  # 커서 객체 생성

cursor.execute("select * from examtable")  # examtable에서 모든 데이터 가져오기
one_rec=cursor.fetchall()  # 가져온 데이터를 한번에 전부 가져옴

for field in one_rec:  # 반복문을 통해 가져온 데이터 하나씩 처리
    kor=field[2]
    eng=field[3] 
    mat=field[4]
    # 각 필드에서 국어, 영어, 수학 점수 추출
    # 각 과목 점수와 총점, 평균 계산 후 출력
    print("이름:%s 학번:%s 국어:%3d 영어:%3d 수학:%3d 총점:%3d 평균:%3d"%(field[0],field[1],kor,eng,mat,kor+eng+mat,int((kor+eng+mat)/3))) 

cursor.close()  # 커서 객체 닫기
con.close()  # 연결된 데이터베이스 닫기
```
