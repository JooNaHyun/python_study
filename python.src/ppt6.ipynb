```python
i=0                #반복 카운터, 초기화 
sum=0              #누적할 변수, 초기화
while i <100 :     #0부터 100까지 실행,만일 실행횟수는 100회 
    i+=1           #i=0+1, i=1+1, i=2+1, i=3+1,i는 1씩 증가 
    sum = sum + i  #sum=0+1, sum=1+2, sum=3+3,sum=6+4 ...
                   #sum은 더한 값이 누적됨
print("sum=",sum)  #while 반복을 빠져 나와서 출력
#이 코드는 1부터 100까지의 합
```

```python
i=1 # i값을 1로 초기화합니다
while True: # 무한 루프를 실행
    j=i #: j값을 i와 같게 초기화
    while True: # 무한 루프를 실행
        if(j==10): break
        #j값이 10이 되면 안쪽 while 루프를 빠져나옴
        print( "*",end="") 
        #별(*)을 출력하고 다음 출력을 이어서 하기 위해 end 옵션에 ""을 지정
        j+=1 #: j값을 1 증가 
    if(i==10): break 
     #i값이 10이 되면 바깥쪽 while 루프를 빠져나옵니다
    print(end="\n") 
    #줄 바꿈
    i+=1
    #i값을 1 증가
```

```python
Mon=1 #Mon을 1로 초기화
for LastDay in [31,28,31,30,31,30,31,31,30,31,30,31]:
#리스트 [31,28,31,30,31,30,31,31,30,31,30,31]를 순회하며 LastDay 변수에 하나씩 할당
    print(Mon,"월 :",end="")
    #현재 월과 :를 출력합니다. end=""는 출력 후 줄바꿈을 하지 않도록 함
    for Day in range(1,LastDay+1):
    #1부터 LastDay+1까지의 범위에서 Day 변수를 순회
        print(Day,end=",") 
        #Day 값을 출력하고 ,로 구분합니다. end=""는 줄바꿈을 하지 않도록 함.
    print(end="\n") #줄바꿈
    Mon+=1 #다음 달로 넘어가기 위해 Mon 값을 1 증가
```

```python
for i in range(2,10,3): # 2단~9단 중 3 간격으로 반복
    for j in range(2,10): # 2~9까지 j 값 반복
        print(i,"*",j,"=",i*j,end="\t") # 구구단 출력한뒤 탭
        print(i+1,"*",j,"=",i*j,end="\t") # 구구단 출력한뒤 탭
        if(i+2!=10): print(i+2,"*",j,"=",i*j,end="\n") 
        ## 3단을 출력하는 경우 ; 구구단 출력한 뒤 탭
        else: print(end="\n")
        # 9*9 구구단 출력 후 개행
    print(end="\n")
    #한 단이 끝나면 다음 줄로 이동
```

```python
for i in range(2,5): # 2에서 4까지 반복하는 i 변수의 for 루프
    for j in range(2,10): #2에서 9까지 반복하는 j 변수의 for 루프
        print(i,"*",j,"=",i*j,end="\t") # i와 j를 곱한 값을 출력하고, 탭으로 분리
        print(i+3,"*",j,"=",i*j,end="\t")# i+3과 j를 곱한 값을 출력하고, 탭으로 분리
        if(i+6!=10): print(i+6,"*",j,"=",i*j,end="\n") 
        # i+6과 j를 곱한 값을 출력하고, 개행
        else: print(end="\n")
        # i+6이 10이 아닐 때만 출력
    print(end="\n")
    #한 번 더 개행
```
